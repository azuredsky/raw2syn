```python

def load_img(path):
        img = cv2.imread(path)
        img = cv2.resize(img, (256, 256))
        img = img/127.5 - 1.
        return img
```

```python


#Root directory of the project
ROOT_DIR = os.path.abspath(".")
MODEL_PATH = os.path.join(ROOT_DIR, "saved_model")
```

```python

g_AB = load_model(os.path.join(MODEL_PATH, 'g_AB.h5'), custom_objects={'InstanceNormalization':InstanceNormalization})
print(g_AB.summary())
```

```python

# Load an image
img = load_img("087.jpg")
print(img.shape)
plt.imshow(img)
```

```python
img_4d = np.expand_dims(img, axis=0)
img_4d.shape
```

```python
syn_img = g_AB.predict(img_4d)
syn_img.shape
```

```python
x = np.squeeze(syn_img, axis=0)
x.shape
```
